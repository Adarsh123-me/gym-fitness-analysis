```python
import pandas as pd

df=pd.read_csv("fitness_datal.csv")
df.head()
```

```python
df.info()
df.describe()
df.isnull().sum()
```

```python
df['Gender']=df['Gender'].str.capitalize() #convert text into this format
df['Workout_Type'] = df['Workout_Type'].str.capitalize()
```

```python
df['Gender'].value_counts() # used for count the gender
```

```python
df.dropna(inplace=True) # use for remove blanks
```

```python
# Total Members 
df['User_ID'].count()
```

```python
# Total Calories
df['Calories_Burned'].sum()
```

```python
# Avg Workout
df['Workout_duration'].mean() #mean used for average value 
```

```python
# Gender Analysis
df.groupby('Gender')['Calories_Burned'].mean()
```

```python
# Workout Type
df.groupby('Workout_type')['Calories_Burned'].sum()
```
