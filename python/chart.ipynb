```python
import pandas as pd
df=pd.read_csv("Fitness_Datal.csv")
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
# used for find the genders quantity
df.groupby('Gender')['Calories_Burned'].mean().plot(kind='pie')
plt.show()
```

```python
# used to find trainer perfomance 
df.groupby('Trainer_Name')['Calories_Burned'].sum().plot(kind='bar') 
plt.show()
```

```python
plt.scatter(df['Steps'], df['Calories_Burned']) # used to find step vs calories 
plt.show()
```
